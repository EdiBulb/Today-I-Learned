# 1.AWS의 개념
## 클라우드 컴퓨팅이란?
### 인터넷을 통해 IT 리소스를 제공하는 서비스
### AWS의 장점 : 유연성, 확장성, 비용 절감, 글로벌 인프라

## AWS 주요 서비스 카테고리
1. 컴퓨팅(EC2, Lambda) : 가상 서버와 서버리스 기능 제공
2. 스토리지(S3, EBS) : 데이터를 저장하고 관리한다
3. 네트워킹(VPC, Route 53): 리소스 간 네트워크 통신 설정
4. 데이터베이스(RDS, DynamoDB) : 관계형 및 비관계형 데이터베이스 제공

# 2.IAM(Identity and Access Management)
역할 : AWS 리소스에 대한 엑세스를 제어한다
주요기능 : 
1. 사용자 및 그룹 생성
2. 정책(Policy)를 사용한 권한 제어
3. MFA(Multi-Factor Authentication) 설정

실습:
- 사용자(User) 생성 및 권한 설정
- 그룹(Group) 생성 후 사용자 추가

# 3.S3(Simple Storage Service)
역할 : 데이터를 저장할 수 있는 객체 스토리지
특징 :
1. 무제한 스토리지
2. 다양한 스토리지 클래스(표준, Glacier 등)
3. 데이터 암호화와 접근 제어 가능

실습 :
- S3 버킷 생성
- 파일 업로드 및 다운로드
- 버킷 정책 설정

# 4. EC2(Elastic Compute Cloud)
역할 : AWS의 가상 서버
특징 : 
- 다양한 인스턴스 유형(CPU, 메모리 조합 선택 가능)
- 온디맨즈, 에약 인스턴스, 스팟 인스턴스 옵션

실습 :
- EC2 인스턴스 생성
- 인스턴스에 연결(SSH)
- 간단한 웹 서버(Apache)설치

# 3주차 : 고가용성 설계
고가용성(High Availability) : 시스템이 장애 상황에서도 지속적으로 서비스를 제공할 수 있는 능력. 즉, 시스템의 가동 시간(Uptime)을 극대화하여, 사용자에게 항상 서비스가 가능하도록 설계된 시스템이다.

왜 고가용성이 중요한가?
1. 서비스 중단 방지
2. 데이터 무결성 유지
3. 고객 경험 향상

고가용성을 구성하는 3요소
1. 장애 내성(Fault Tolerance)
2. 장애 복구(Disaster Recovery)
3. 확장성(Scalability)

고가용성을 달성하기 위한 AWS 도구
1. 다중 가용 영역(Multi-AZ)
- AWS는 여러 가용 영역(AZ, Availability Zone)을 제공한다
- 고가용성을 위해 리소스를 여러 AZ에 분산 배치한다.
- 예: RDS Multi-AZ로 데이터베이스를 하나의 AZ가 아닌 다른 AZ에도 복제한다.

2. Auto-Scaling
- 트래픽 증가 시 EC2 인스턴스를 자동으로 추가하고, 트래픽 감소 시 제거하여 항상 적절한 리소스를 유지한다.

3. Elastic Load Balancer(ELB)
- 트래픽을 여러 인스턴스에 분산해 부하를 줄이고, 하나의 인스턴스가 다운되더라도 다른 인스턴스가 요청을 처리한다.

4. Route 53 헬스 체크
- 특정 리소스가 비정상적으로 작동하면 다른 리소스로 트래픽을 전환

5. 백업 및 복구
- AWS Backup, S3 버전 관리, Glacier를 사용해 데이터를 정기적으로 백업

6. CloudFront
- 엣지 로케이션에서 캐싱된 데이터를 제공하여 사용자 요청을 빠르게 처리하고, 장애 발생 시에도 콘텐츠 제공한다


예시: 고가용성 아키텍쳐
- 웹 애플리케이션
1. 웹 서버(EC2)를 두 개 이상의 AZ에 배치
2. ELB로 트래픽을 균등하게 분산
3. 데이터베이스는 Multi-AZ RDS로 복제
4. S3를 사용해 정적 콘텐츠 제공
5. Auto Scaling으로 트래픽 증가에 대비


고가용성을 위한 설계 원칙
1. 중복성(Redundancy)
- 모든 주요 구성 요소를 이중화(Backup)하여 하나가 실패해도 다른 하나가 작동
- 예: 두 개의 데이터베이스를 서로 다른 AZ에 배치
2. 장애 포인트 최소화
- 시스템의 단일 장애 지점(Single point of Failure, SPOF)을 제거
- 예: 단일 서버 대신 다중 서버 구성
3. 모니터링 및 자동화
- CloudWatch를 사용해 상태를 모니터링하고, 문제가 발생하면 자동 복구
