# 3주차 : 고가용성 설계
고가용성(High Availability) : 시스템이 장애 상황에서도 지속적으로 서비스를 제공할 수 있는 능력. 즉, 시스템의 가동 시간(Uptime)을 극대화하여, 사용자에게 항상 서비스가 가능하도록 설계된 시스템이다.

왜 고가용성이 중요한가?
1. 서비스 중단 방지
2. 데이터 무결성 유지
3. 고객 경험 향상

고가용성을 구성하는 3요소
1. 장애 내성(Fault Tolerance)
2. 장애 복구(Disaster Recovery)
3. 확장성(Scalability)

고가용성을 달성하기 위한 AWS 도구
1. 다중 가용 영역(Multi-AZ)
- AWS는 여러 가용 영역(AZ, Availability Zone)을 제공한다
- 고가용성을 위해 리소스를 여러 AZ에 분산 배치한다.
- 예: RDS Multi-AZ로 데이터베이스를 하나의 AZ가 아닌 다른 AZ에도 복제한다.

2. Auto-Scaling
- 트래픽 증가 시 EC2 인스턴스를 자동으로 추가하고, 트래픽 감소 시 제거하여 항상 적절한 리소스를 유지한다.

3. Elastic Load Balancer(ELB)
- 트래픽을 여러 인스턴스에 분산해 부하를 줄이고, 하나의 인스턴스가 다운되더라도 다른 인스턴스가 요청을 처리한다.

4. Route 53 헬스 체크
- 특정 리소스가 비정상적으로 작동하면 다른 리소스로 트래픽을 전환

5. 백업 및 복구
- AWS Backup, S3 버전 관리, Glacier를 사용해 데이터를 정기적으로 백업

6. CloudFront
- 엣지 로케이션에서 캐싱된 데이터를 제공하여 사용자 요청을 빠르게 처리하고, 장애 발생 시에도 콘텐츠 제공한다


예시: 고가용성 아키텍쳐
- 웹 애플리케이션
1. 웹 서버(EC2)를 두 개 이상의 AZ에 배치
2. ELB로 트래픽을 균등하게 분산
3. 데이터베이스는 Multi-AZ RDS로 복제
4. S3를 사용해 정적 콘텐츠 제공
5. Auto Scaling으로 트래픽 증가에 대비


고가용성을 위한 설계 원칙
1. 중복성(Redundancy)
- 모든 주요 구성 요소를 이중화(Backup)하여 하나가 실패해도 다른 하나가 작동
- 예: 두 개의 데이터베이스를 서로 다른 AZ에 배치
2. 장애 포인트 최소화
- 시스템의 단일 장애 지점(Single point of Failure, SPOF)을 제거
- 예: 단일 서버 대신 다중 서버 구성
3. 모니터링 및 자동화
- CloudWatch를 사용해 상태를 모니터링하고, 문제가 발생하면 자동 복구



# 확장성 설계
- 트래픽 증가에도 안정적으로 서비스를 제공할 수 있는 확장 가능한 아키텍쳐를 설게한다.
- AWS의 서비스 간 연동을 통해 효율적인 워크로드 처리 방식을 학습한다.

1. CloudFront
역할: 글로벌 CDN(Content Delivery Network)으로, 사용자에게 콘텐츠를 빠르게 전달
특징: 엣지 로케이션(Edge Location)에서 콘텐츠를 캐싱해 빠른 응답 제공
실습: S3에 호스팅된 정적 웹사이트를 CloudFront와 연결해 속도 향상

2. DynamoDB
역할: 서버리스 NoSQL 데이터베이스로, 대량의 데이터를 빠르게 처리
특징: 완전 관리형, 고확장성, 자동 확장(Auto Scaling)지원
실습: DynamoDB 테이블 생성 및 애플리케이션과 연결

3. S3 버킷의 확장성 활용
역할: 무제한 확장 가능한 스토리지
사용사례:
- 정적 파일(이미지, 동영상) 호스팅
- 데이터 백업 및 아카이빙

실습:
- S3에 정적 웹사이트 호스팅
- S3 이벤트 트리거를 사용해 데이터 처리 워크플로우 실행한다.

4. Amazon SQS (Simple Queue Service)
- 역할: 메시지 큐를 통해 비동기적으로 워크로드 처리
- 사용 사례: 애플리케이션 간 메시지 전달 및 처리 지연 허용
실습: SQS를 사용해 주문 처리 워크플로우 구성

5. Amazon SNS(Simple Notification Service)
- 역할: 알림 서비스를 통해 이메일, 문자, 푸시 알림을 전송
- 사용 사례:
특정 이벤트 발생 시 알림 전송
- 실습:
SNS 토픽 생성 및 이메일 구독 설정




# 5~6주차 : AWS 보안와 비용 최적화
## 5주차 : 보안 설계
학습 목표
- AWS 리소스를 안전하게 보호하는 방법을 학습한다.
- 데이터 암호화, 권환 관리, 모니터링 등을 통해 AWS 환경을 강화한다.

1. IAM(Identity and Access Management)심화
- 권한 정책(Policy): JSON 기반의 정책을 작성하여 세부적인 권한을 제어한다.
- 역할(Role): 애플리케이션이나 AWS 서비스에 필요한 권한을 부여한다.
- MFA(Multi-Factor Authentication): 이중 인증 설정으로 보안 강화한다.
실습
- IAM 정책 작성 및 사용자 권한 제한
- 역할(Role)을 생성하여 EC2가 S3에 접근하도록 설정한다.

2. S3 데이터 보안
- 버킷 정책: 특정 사용자나 서비스만 버킷에 접근하도록 제어한다.
- 서버 측 암호화(Server-Side Encryption): AWS KMS(Key Management Service)를 사용한 암호화
- 버전 관리(Versioning): 실수로 데이터가 삭제되거나 덮어씌워지는 것을 방지한다.
실습
- S3 버킷에 접근 정책 설정
- KMS를 사용해 데이터를 암호화 및 복호화

3. AWS KMS(Key Management Service)
- 역할: 암호화 키를 생성 및 관리하여 데이터 암호화
- 사용사례: RDS, S3, DynamoDB 데이터 암호화
실습
- KMS 키 생성 및 관리
- S3 데이터를 KMS로 암호화

4. CloudTrail
- 역할: AWS계정에서의 API 호출 및 활동을 로깅
- 사용 사례: 누가 어떤 리소스에 접근했는지 감사(Audit)
- 실습: CloudTrail 활성화 및 로그 분석

실습
- S3 버킷을 Glacier로 이동
- S3 Intelligent-Tiering 활성화

4. EC2 비용 최적화
